Importing dependencies


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the Boston House Price Dataset

```python
housing = fetch_california_housing(as_frame=True)
df = housing.frame
```

```python
print(housing)
```

```python
housing.frame['MedHouseVal']=housing.target
```

```python
housing.frame.head()
```

```python
housing.frame.shape
```

```python
#check for missing values
housing.frame.isnull().sum()
```

```python
#statistical measures of the dataset
housing.frame.describe()
```

Understanding the correlation between various fearures in the dataset

```python
correlarion = housing.frame.corr()
```

```python
#constructing heatmap to understand correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlarion,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

Splitting the data and target

```python
X= housing.frame.drop(columns='MedHouseVal',axis=1)
Y= housing.frame['MedHouseVal']
```

```python
print(X)
print(Y)
```

Splitting the data into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

XGBoost Regressor

```python
#loading the model
model = XGBRegressor()
```

```python
# training the model with x_train
model.fit(X_train,Y_train)
```

Evaluation

Prediction on training data

```python
#accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train,training_data_prediction)

#mean absolute
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)

print("R squared error : ",score_1)
print("Mean absolute error : ",score_2)
```

Visualize the actual and predicted values

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

Prediction on test data

```python
#accuracy for prediction on test data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test,test_data_prediction)

#mean absolute
score_2 = metrics.mean_absolute_error(Y_test,test_data_prediction)

print("R squared error : ",score_1)
print("Mean absolute error : ",score_2)
```

```python
#plot
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
